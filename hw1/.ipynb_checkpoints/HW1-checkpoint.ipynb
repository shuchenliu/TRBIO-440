#### 3. Suppose that the weights of livers from rats treated with a toxic substance is normally distributed with a mean μ = 44g and a standard deviation of σ = 5g. Calculate the Z scores and probabilities for:

**a.**  a liver from a rat of this population having a weight between 43g and 45g.

```r
z45 <- (45 - 44) / 5 
z45
```

```r
z43 <- (43 - 44) / 5
z43
```

```r
prob <- pnorm(z45) - pnorm(z43) 
prob
```

**b**.  a liver from a rat of this population having a weight of less than 42g.

```r
z42 <- (42 - 44) / 5
z42
```

```r
prob <- pnorm(z42)
prob
```

**c.** a liver from a rat of this population having a weight of greater than 52g.

```r
z52 <- (52 - 44) / 5
z52
```

```r
prob <- pnorm(z52, lower.tail=FALSE)
prob
```

#### 4. For the same distribution (mean μ = 44g and a standard deviation of σ = 5g), give the liver weights that define the following: [8 pts.]

 **a**. 73rd percentile of the distribution

```r
getPercentileValue = function(p) {
    return (qnorm(p, mean=44, sd=5))
}

getPercentileValue(0.73)
```

**b.**  the middle 95th percent of the distribution (range of values)

```r
delta = (1 - 0.95) / 2;
upperLimit <- getPercentileValue(0.95 + delta)
upperLimit
```

```r
lowerLimit <- getPercentileValue(delta)
lowerLimit
```

**c.**  43% of the values in the right tail of the distribution.



```r
left <- 1 - .43;
getPercentileValue(left)
```
