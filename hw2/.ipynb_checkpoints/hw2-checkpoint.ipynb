### 1. We have the following observed measurements of percent calcium in teeth from 10 individuals:   36.39, 36.19, 34.20, 35.15, 35.47, 35.22, 36.11, 35.63, 36.63, 35.59 [15 pts]

#### a. Graph the data.

```r
data <- c(36.39, 36.19, 34.20, 35.15, 35.47, 35.22, 36.11, 35.63, 36.63, 35.59)
hist(data, 
     main="Measurement of Percent Calcium in Teeth", 
     xlab="percentage", 
     xlim=c(34,37),
     ylim=c(0,3),
     yaxt="n",
     col="lightblue", 
     breaks=5)
axis(2, at=0:3)
```
